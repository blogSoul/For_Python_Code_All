```python
def EUC(u, v):
    A =  np.sum(np.square(u - v))
    B = np.sqrt(A)
    return B
#두 리스트에 대한 유클리드 거리입니다.

def EUC_array(a):
    number_of_user = np.size(a, axis = 0)
    N = np.full((number_of_user, number_of_user), 0.0)
    #배열 a의 크기에 맞게 빈 배열을 만듭니다.
    for u in range(0, number_of_user):
        for v in range(u, number_of_user):
            A =  np.sum(np.square(a[u,] - a[v,]))
            N[v, u] = np.sqrt(A)
            N[u, v] = N[v, u]
    return N
#하나의 배열에 대한 유클리드 거리입니다.
```

```python
def JAC(u, v):
    U = np.array(u > 0,dtype = np.int)
    V = np.array(v > 0, dtype = np.int)
    N = U + V
    inter = np.sum(np.array(N > 1, dtype = np.int)) # 배열의 교집합 부분입니다. 
    union = np.sum(np.array(N > 0, dtype = np.int)) # 배열의 합잡합 부분입니다.
    return inter/union
#두 리스트에 대한 JAC를 구하는 알고리즘입니다.

def JAC_array(a):
    number_of_user = np.size(a, axis = 0)
    N = np.full((number_of_user, number_of_user), 0.0)
    #배열 a의 크기에 맞게 빈 배열을 만듭니다.
    for u in range(0, number_of_user):
        for v in range(0, number_of_user):
            m = JAC(a[u, ],a[v, ])
            N[u, v] = m
            N[v,u] = N[u, v]
    return N
#하나의 배열에 대한 AC를 구하는 알고리즘입니다.
```

```python
def MSD(u, v):
    U = np.where(u == 0, np.nan, u)
    V = np.where(v == 0, np.nan, v)
    sum = np.square(U - V)
    sum = sum[~np.isnan(sum)]
    
    size = np.size(sum, axis = 0)
    
    N = np.sum(sum)/size
    return N
#두 리스트에 대한 MSD를 구하는 알고리즘입니다.

def MSD_array(a):
    number_of_user = np.size(a, axis = 0)
    N = np.full((number_of_user, number_of_user), 0.0)
    #배열 a의 크기에 맞게 빈 배열을 만듭니다.
    for u in range(0, number_of_user):
        for v in range(0, number_of_user):
            m = MSD(a[u,],a[v,])
            N[u,v] = m
            N[v,u] = N[u, v]
    return N
#하나의 배열에 대한 MSD를 구하는 알고리즘입니다.
```

```python
def JMSD(u, v, max):
    return JAC(u, v)* (1 - (MSD(u/max, v/max)))
#두 리스트에 대한 JMSD를 구하는 알고리즘입니다.

def JMSD(a, max):
    return JAC(a) * (1 - (MSD(a/max)))
#하나의 배열에 대한 JMSD를 구하는 알고리즘입니다.
```

```python
def COS(u,v):
    return np.dot(u,v)/(np.linalg.norm(u) * np.linalg.norm(v))
#두 리스트에 대한 COS를 구하는 알고리즘입니다.

def COS_array(a):
    number_of_user = np.size(a, axis = 0)
    N = np.full((number_of_user, number_of_user), 0.0)
    #배열 a의 크기에 맞게 빈 배열을 만듭니다.
    for u in range(0, number_of_user):
        for v in range(0, number_of_user):
            m = COS(a[u, ],a[v, ])
            N[u,v] = m
            N[v,u] = N[u, v]
    return N
#하나의 배열에 대한 COS를 구하는 알고리즘입니다.
```

```python
def COS_zero(u,v):
    zero_u = np.where(u== 0)
    zero_v = np.where(v ==0)
    zero = np.concatenate((zero_u, zero_v),axis = None)
    
    U = np.delete(u, zero)
    V = np.delete(v, zero)
    return np.dot(U, V)/(np.linalg.norm(U) * np.linalg.norm(V))
#두 리스트(NULL 부분 제외)에 대한 COS를 구하는 알고리즘입니다.

def COS_zero_array(a):
    number_of_user = np.size(a, axis = 0)
    N = np.full((number_of_user, number_of_user), 0.0)
    #배열 a의 크기에 맞게 빈 배열을 만듭니다.
    for u in range(0, number_of_user):
        for v in range(0, number_of_user):
            m = COS_zero(a[u, ],a[v, ])
            N[u,v] = m
            N[v,u] = N[u, v]
    return N
#하나의 배열(NULL 부분 제외)에 대한 COS를 구하는 알고리즘입니다.
```

```python
def PCC(u, v):
    U_mean = np.nanmean(np.where(u!=0,np.nan))
    V_mean = np.nanmean(np.where(v!=0,np.nan))
    
    zero_u = np.where(u == 0)
    zero_v = np.where(v == 0)
    zero = np.concatenate((zero_u, zero_v), axis =None)
    
    U = np.delete(u, zero_u)
    V = np.delete(v, zero_v)
    U = U - U_mean
    V = V - V_mean
    
    return np.dot(U, V)/(np.linalg.norm(U) * np.linalg.norm(V))
#두 리스트(NULL 부분 제외)에 대한 PCC를 구하는 알고리즘입니다.

def PCC_array(a):
    number_of_user = np.size(a, axis = 0)
    N = np.full((number_of_user, number_of_user), 0.0)
    #배열 a의 크기에 맞게 빈 배열을 만듭니다.
    mean = np.nanmean(np.where(a!=0, a, np.nan), axis=1)
    
    for u in range(0, number_of_user):
        for v in range(0, number_of_user):
            zero_u = np.where(a[u, ] == 0)
            zero_v = np.where(a[v, ] == 0)
            zero = np.concatenate((zero_u, zero_v), axis =None)
      
            U = np.delete(a[u, ], zero)
            V = np.delete(a[v, ], zero)
            U = U - mean[u]
            V = V - mean[v]
            
            Inner = np.dot(U,V)
            N[u, v] = Inner/(np.linalg.norm(U) * np.linalg.norm(V))
            N[v,u] = N[u, v]
    return N
#하나의 배열(NULL 부분 제외)에 대한 PCC를 구하는 알고리즘입니다.
```

```python
def CPCC(u, v):
    U_median = np.nanmedian(np.where(u!=0,u,np.nan))
    V_median = np.nanmedian(np.where(v!=0,v,np.nan))
    
    zero_u = np.where(u == 0)
    zero_v = np.where(v == 0)
    zero = np.concatenate((zero_u, zero_v), axis =None)
    
    U = np.delete(u, zero)
    V = np.delete(v, zero)
    U = U - U_median
    V = V - V_median
    
    return np.dot(U, V)/(np.linalg.norm(U) * np.linalg.norm(V))
#두 리스트(NULL 부분 제외)에 대한 CPCC를 구하는 알고리즘입니다.
# * 평균이 아니라 중간값을 사용합니다.

def CPCC_array(a):
    number_of_user = np.size(a, axis = 0)
    N = np.full((number_of_user, number_of_user), 0.0)
    #배열 a의 크기에 맞게 빈 배열을 만듭니다.
    median = np.nanmedian(np.where(a!=0, a, np.nan), axis=1)
    
    for u in range(0, number_of_user):
        for v in range(0, number_of_user):
            zero_u = np.where(a[u, ] == 0)
            zero_v = np.where(a[v, ] == 0)
            zero = np.concatenate((zero_u, zero_v), axis =None)
      
            U = np.delete(a[u, ], zero)
            V = np.delete(a[v, ], zero)
            U = U - median[u]
            V = V - median[v]
            
            Inner = np.dot(U,V)
            N[u, v] = Inner/(np.linalg.norm(U) * np.linalg.norm(V))
            N[v,u] = N[u, v]
            N[np.isnan(N)] = -1
    return N    
```
